```python
cate = "primaryNivo"
edge_os = pd.read_csv(f"./cox_result/cox_result_{cate}_os.txt",index_col=0)
edge_pfs = pd.read_csv(f"./cox_result/cox_result_{cate}_pfs.txt",index_col=0)


ano = ".positive."
resos = pd.read_csv(f"../connectivity/{cate}{ano}_os_significant_nodes_by_connectivity.txt")
respfs = pd.read_csv(f"../connectivity/{cate}{ano}_pfs_significant_nodes_by_connectivity.txt")

ano = ".negative."
resos1 = pd.read_csv(f"../connectivity/{cate}{ano}_os_significant_nodes_by_connectivity.txt")
respfs1 = pd.read_csv(f"../connectivity/{cate}{ano}_pfs_significant_nodes_by_connectivity.txt")

#degree_os = pd.read_csv(f"../degree/pn_os_significant_nodes_by_degree.txt")
#degree_pfs = pd.read_csv(f"../degree/pn_os_significant_nodes_by_degree.txt")
```

```python
p = 0.01
a = [ j for i in edge_os[edge_os['p']<p ].index.tolist() for j in i.split("_")]
b = [ j for i in edge_pfs[edge_pfs['p']<p ].index.tolist() for j in i.split("_")]
c = resos[resos['p']<p ]["covariate"].tolist() + resos1[resos1['p']<p ]["covariate"].tolist()
d = respfs[respfs['p']<p ]["covariate"].tolist() + respfs1[respfs1['p']<p ]["covariate"].tolist()

#e = degree_os[degree_os['p']<p ]["covariate"].tolist()
#f = degree_pfs[degree_pfs['p']<p ]["covariate"].tolist()
```
