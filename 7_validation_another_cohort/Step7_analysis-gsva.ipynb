```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np 
import seaborn as sns

sns.set_context("notebook", rc={"lines.linewidth": 3})
##################################################
plt.style.use('tableau-colorblind10')
##################################################
```

```python
dfcl = pd.read_csv("../motzer2021_clinical_data.txt",sep="\t")
cate = "gsva" # edge_entropy.all eigenvector closeness edge_betweenness
dfpa = pd.read_csv(f"gsva_c2kegg_pathway_354-pd-L1.txt",sep="\t")
```

```python
path_list =  dfpa.index.tolist()
```

```python
dfpa.T.reset_index()
```

```python
dfmerge = pd.merge(dfpa.T.reset_index(),dfcl,how="left",left_on="index",right_on="ID")
dfmerge.columns = ["index"] + dfmerge.columns.tolist()[1:]
dfmerge
```

```python
from lifelines import CoxPHFitter
df_cox = pd.DataFrame()
for i in  path_list:
    try:
        #print(i)
        df_pfs = dfmerge[[i,'PFS_P', 'PFS_P_CNSR']].copy()
        cph = CoxPHFitter()
        cph.fit(df_pfs, duration_col='PFS_P', event_col='PFS_P_CNSR')#,show_progress=True)
        #print(i, cph.summary,"-------------")
        df_cox = pd.concat([df_cox,cph.summary])
    except:
        pass

df_cox
```

```python
colors = dfmerge.set_index("index")["PFS_P"]
#print(colors)
colors1 = []#[ "blue" if i >8 elif "red" for i in colors]
for i in colors:
    if i >6:
        colors1.append("blue")
    else:
        colors1.append("red")
colors = pd.DataFrame(colors1,index=dfmerge["index"],columns=["PFS"])
colors
```

```python
import math
def secline(alist):
    newlist = ""
    if len(alist) > 20:
        if len(alist.split("_")) > 2: 
            num = math.floor(len(alist.split("_"))/2)
            newlist += "_".join(alist.split("_")[:num])
            newlist +="\n"+ "_".join(alist.split("_")[num:])
    else:
        newlist = alist
    return newlist
```

```python
pfs_list = df_cox[df_cox["p"]<0.05].index.tolist()
```

```python
import seaborn as sns
import scienceplots
import matplotlib.pyplot as plt
import matplotlib.colors

final_list = pfs_list[:]

dfpasp = dfpa[final_list]
dfpasp.columns = [secline(i) for i in dfpasp.columns.tolist()]
g1 = sns.clustermap(dfpasp,
               method="ward",metric="cosine",cmap=sns.color_palette("coolwarm", 7),
               row_colors= colors,col_cluster=False,
               row_cluster=True,yticklabels=False,z_score=1,figsize=(5,8))#,cmap="Blues")

plt.setp(g1.ax_heatmap.get_xticklabels(), rotation=90) 
ax = g1.ax_heatmap
ax.set_ylabel("")
g1.ax_cbar.set_position((1, .2, .03, .4))


dic = {"pfs<=6":"red","pfs>6":"blue"}

for label in ["pfs<=6","pfs>6"]:
    g1.ax_col_dendrogram.bar(0, 0, color=dic[label], label=label, linewidth=0)
l1 = g1.ax_col_dendrogram.legend(title='samples', loc="best", ncol=1, bbox_to_anchor=(1.35, 0.10))
"""
dic1 = {1:"blue",2:"orange",3:"green",4:"red"}
for label in [1,2,3,4]:
    g1.ax_col_dendrogram.bar(0, 0, color=dic1[label], label=f"c{label}", linewidth=0)
l1 = g1.ax_col_dendrogram.legend(title='samples', loc="best", ncol=1, bbox_to_anchor=(1.15, 0.20))
"""
plt.savefig(f"f6.motzer_path_{cate}.svg",dpi=600,bbox_inches='tight')
plt.show()
plt.close()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
from scipy.stats import zscore
from scipy.spatial.distance import pdist

linkage_data = linkage(pdist(zscore(
    dfpa[final_list],axis=0), 'cosine'),
                       method='ward', metric='euclidean')
dn = dendrogram(linkage_data)
plt.show() 
```

```python
from lifelines import KaplanMeierFitter
from lifelines.datasets import load_waltons
from sksurv.compare import compare_survival
import numpy as np
from scipy.cluster.hierarchy import fcluster

fl = fcluster(linkage_data,2,criterion='maxclust')
#print(fl)
df_os = dfmerge.copy()

def swapPositions(lis):
    newlis = []
    for i in lis:
        if i == 1:
            newlis.append(2)
        if i == 2:
            newlis.append(1)
    return newlis

df_os["cres"] = fl
df_os["cres"] =swapPositions(fl)
```

```python
y = df_os[["PFS_P_CNSR","PFS_P"]].values
y = np.array([ (i[0],i[1]) for i in y],dtype="bool,f")

#df_os_fl[df_os_fl[tar]>df_os[tar].quantile(0.65)] = "higher"
group_indicator =  fl
chi2, pvalue= compare_survival(y,group_indicator)
print(pvalue)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sksurv.nonparametric import kaplan_meier_estimator
dc = {2:"#e40c69",1:"#2e3092",3:'#6d6e71',4:"black"}
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(4.5, 2))

df_os["PFS_P_CNSR"] = df_os["PFS_P_CNSR"]==1

for i in set(fl):
    speci_df = df_os[df_os["cres"]==i]

    time, survival_prob, conf_int = kaplan_meier_estimator(
        speci_df["PFS_P_CNSR"], \
        speci_df["PFS_P"], conf_type="log-log"
)
    plt.step(time, survival_prob, where="post", label=f"c{i}",color=dc[i])
    #plt.fill_between(time, conf_int[0], conf_int[1], alpha=0.25, step="mid")
    
ax.text(0.1, 0.3, "p = "+str(round(pvalue,5)), transform=ax.transAxes, fontsize=14,
        verticalalignment='top')

plt.ylabel("Survival probability")
plt.xlabel("Time:month (PFS)")
plt.legend(loc="best")  
plt.title("Cloesness")
#plt.title(f"{cate} scores")
plt.savefig(f"f6.motzer_{cate}.survival.svg",dpi=600,bbox_inches='tight')
plt.show()
```
