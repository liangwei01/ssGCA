```python
df_cl = pd.read_csv("../leave_one_out/braun_data_clinical_data.csv")
df_cl.head()
```

```python
dfexp = pd.read_csv(f"../leave_one_out/Expression_log2_21_{cate}_braun_2020.csv",index_col=0)
dfexp.loc[:,interlist].reset_index()
```

```python
df_merge = pd.merge(dfexp.loc[:,interlist].reset_index(),df_cl,how='left', left_on="gene_name", right_on="RNA_ID" )
```

```python
from lifelines import CoxPHFitter
df_cox = pd.DataFrame()

for i in interlist:
    try:
        df_pe1 = df_merge[[i,'OS', 'OS_CNSR']].copy()
        df_pe1[i] = df_pe1[i] /10
        cph = CoxPHFitter()
        cph.fit(df_pe1, duration_col='OS', event_col='OS_CNSR')
        df_cox = pd.concat([df_cox,cph.summary])
        #print(cph.summary)
    except:
        pass
    #break
```

```python
df_cox[df_cox["p"]<0.05].index
```

```python
from lifelines import CoxPHFitter
df_cox1 = pd.DataFrame()

for i in interlist:
    try:
        df_pe1 = df_merge[[i,'PFS', 'PFS_CNSR']].copy()
        df_pe1[i] = df_pe1[i] /10
        cph = CoxPHFitter()
        cph.fit(df_pe1, duration_col='PFS', event_col='PFS_CNSR')
        df_cox1 = pd.concat([df_cox1,cph.summary])
        #print(cph.summary)
    except:
        pass
    #break
```
