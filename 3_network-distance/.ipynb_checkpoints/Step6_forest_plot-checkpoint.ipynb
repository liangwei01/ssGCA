```python
import forestplot as fp
import pandas as pd
#https://pypi.org/project/forestplot/
```

```python
from matplotlib import pyplot as plt
import scienceplots
import seaborn as sns
```

```python
df_cl = pd.read_csv("../braun_data_log2_or_not/braun_data_clinical_data.csv")

for sampleGroup in ["metaNivo"]:#, "primaryEver", "metaNivo","metaEver"]:
    df_nd = pd.read_csv(f"all_sample_{sampleGroup}_edge_number.r_filter.txt",sep=" ",header=None)
    df_os = pd.merge(df_nd,df_cl,how='left', left_on=0, right_on="RNA_ID")
    df_os.sort_values(by="Benefit",inplace=True)

    fos = open(f"cph_os_{sampleGroup}.txt","w+")
    ffps = open(f"cph_fps_{sampleGroup}.txt","w+")
    for j in [1,2,3]:
        if j == 1:
            df_os["nd"] = df_os[1] *100
        elif j ==2:
            df_os["nd"] = (df_os[2] ) *100
        elif j ==3:
            df_os["nd"] = (df_os[1] - df_os[2] ) *100
            
        from lifelines import CoxPHFitter
        df_os1 = df_os[["nd",'OS', 'OS_CNSR']]
        cph = CoxPHFitter()
        cph.fit(df_os1, duration_col='OS', event_col='OS_CNSR')
        print("-------OS --------",cph.summary)
        
        fos.write(sampleGroup+"\t"+"\t".join([str(i) for i in cph.summary.values.tolist()[0]])+"\n")
        df_os1 = df_os[["nd",'PFS', 'PFS_CNSR']]
        cph = CoxPHFitter()
        cph.fit(df_os1, duration_col='PFS', event_col='PFS_CNSR')
        print("-------PFS --------",cph.summary)
        ffps.write(sampleGroup+"\t"+"\t".join([str(i) for i in cph.summary.values.tolist()[0]])+"\n")
    
fos.close()
ffps.close()
```

```python
header_txt = pd.read_csv("header.txt",sep=",")
header_txt.columns.tolist()
```

```python
sampleGroup = "metaNivo"

dfos = pd.read_csv(f"cph_os_{sampleGroup}.txt",sep="\t",header=None)
dfpfs = pd.read_csv(f"cph_fps_{sampleGroup}.txt",sep="\t",header=None)
df = pd.concat([dfos,dfpfs])
df.columns = header_txt.columns.tolist()
df["covariate"] = ["Nd_CB", "Nd_NCB","Nd_CB2NCB","Nd_CB", "Nd_NCB","Nd_CB2NCB"]
df["group"] = 3 * ["OS"] + 3* ["PFS"]
df 
```

```python
#plt.rc('font', family='DejaVu Sans')

fp.forestplot(df,  # the dataframe with results data
              estimate="exp(coef)",  # col containing estimated effect size 
              ll="exp(coef) lower 95%", hl="exp(coef) upper 95%",  # columns containing conf. int. lower and higher limits
              varlabel="covariate",  # column containing variable label
              annote=["est_ci"],  # columns to report on left of plot
              annoteheaders=["Est. (95% Conf. Int.)"],
              groupvar="group",  # Add variable groupings  # sort in ascending order (sorts within group if group is specified)               
              pval="p",  # Column of p-value to be reported on right
              xlabel="hazzard ratio",
              table=True, 
            **{"ylabel1_size": 11,
                "marker": "D",  # set maker symbol as diamond
                 "markersize": 35,  # adjust marker size
                 "xlinestyle": (1, (10, 5)),  # long dash for x-reference line 
                 "xlinecolor": "#808080",  # gray color for x-reference line
                 "xtick_size": 12,  # adjust x-ticker fontsize
               } 
              )
plt.tight_layout()
plt.gcf().set_size_inches(4, 3)
plt.savefig(f"{sampleGroup}_forest_plot.tif", dpi=600, bbox_inches='tight')
plt.savefig(f"{sampleGroup}_forest_plot.svg", dpi=600, bbox_inches='tight')
plt.show()
```
